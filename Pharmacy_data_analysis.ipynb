### Removing Columns all null values

```python
data.dropna(axis = 1,how ='all')
```

### Removing rows with all null values

```python
data.dropna(axis = 0,how = 'all')
```

```python
data_int = data.select_dtypes(include=[int])
data_int
```

### checking the columns which has zeroes in all the rows

```python
values = data_int.sum()
values
```

```python
data_int.keys()
```

```python
zero_columns = []
for i in data_int.keys():
    if values[i]==0:
        zero_columns.append(i)
print(zero_columns)
```

### Dropping Columns which has all zeroes

```python
cleaned_data = data.drop(columns = zero_columns)
```

```python
cleaned_data.head()
```

### Replacing the spaces with underscore( _ ) so that we can insert our dataframe in mySQL effectively

```python
cleaned_data.columns = cleaned_data.columns.str.replace(' ','_')
```

```python
cleaned_data['State'].value_counts()
```

```python
cleaned_data['State'] = cleaned_data['State'].replace(1,'Tamil Nadu')
```

```python
cleaned_data['State'] = cleaned_data['State'].fillna('Tamil Nadu')
cleaned_data.head()
```

```python
cleaned_data.isnull().sum()
```

### columns with more null values

```python
null_count = cleaned_data.isnull().sum()
mean = 3002
null_count_list = []

for i in null_count.keys():
    if null_count[i]>3002:
        null_count_list.append(i)
        
print(null_count_list) 
```

### Dropping the columns with highest null values

```python
cleaned_data = cleaned_data.drop(columns = null_count_list)
```

```python
cleaned_data.head()
```

```python
cleaned_data['Bill_Date'] = pd.to_datetime(cleaned_data['Bill_Date'],format ='%d/%m/%Y')
```

```python
cleaned_data['Month_year'] = cleaned_data['Bill_Date'].dt.strftime('%Y %B')
```

```python
cleaned_data.head()
```

```python
cleaned_data.info()
```

```python
cleaned_data['Bill_Date']
```

```python
new_data = cleaned_data[['Bill_Date','Month_year','Bill_Amount']]
new_data.head()
```

### Calculating Monthly sales 

```python
monthly_sales = new_data.groupby('Month_year')['Bill_Amount'].sum().reset_index()
monthly_sales = monthly_sales.sort_values('Month_year')
```

```python
monthly_sales
```

### Calculating monthly sales and tax

```python
sales = cleaned_data.groupby('Month_year')[['Bill_Amount','Goods_Value','SGST_Amount','CGST_Amount','Total_Tax.Amt']].sum().reset_index()
sales.head()
```

### Calculating GST

```python
sales['GST'] = sales['SGST_Amount'] + sales['CGST_Amount']
```

```python
sales['Date'] = pd.to_datetime(sales['Month_year'], format='%Y %B')
sales
```

```python
sales = sales.sort_values('Date')
```

```python
#sales['profit'] = sales['Bill Amount'] - sales['Goods Value']
sales['Month_year']
```

```python
sales.to_excel(r'C:\Users\srira\Downloads\A.R medical data\A.R medical Sales.xlsx')
```

### Analyzing Monthly Sales: Line Chart Visualization

```python
plt.figure(figsize=(20,7))
plt.plot(sales['Month_year'],sales['Bill_Amount'],marker = 'o',linestyle = "solid",markerfacecolor = 'red',markersize = 10)
```

### Analyzing Monthly Sales: Plot and Bar Chart Visualization

```python
plt.figure(figsize=(20,7))
plt.title('Sales over Months',fontweight='bold')
plt.xlabel('Months',fontweight='bold')
plt.ylabel('Sales',fontweight='bold')
plt.plot(sales['Month_year'],sales['Bill_Amount'],marker = 'o',linestyle = "solid",markerfacecolor = 'red',markersize = 10)
plt.bar(sales['Month_year'],sales['Bill_Amount'],color = 'green')
```

### Piechart

```python
plt.figure(figsize=(7,7))
plt.pie(list(sales['Bill_Amount']),labels = list(sales['Month_year']),autopct='%0.1f%%')
plt.show()
```

```python
customer = cleaned_data.groupby('Customer_Name')[['Bill_Amount','Cash_Disc.Amt']].sum().reset_index()
customer = customer.sort_values('Bill_Amount',ascending= False)
customer
```

```python
plt.figure(figsize=(17,5))
plt.title('Customers with highest purchase rate',fontsize=15,fontweight='bold')
plt.xlabel('Customer Names',fontsize=15,fontweight='bold')
plt.ylabel('Amount',fontsize=15,fontweight='bold')
plt.bar(customer['Customer_Name'][0:10],customer['Bill_Amount'][0:10],color = 'darkgreen')
```

```python
discount = customer.sort_values('Cash_Disc.Amt',ascending = False)
plt.figure(figsize=(20,8))
plt.title('Customers with highest discount rate',fontsize=20,fontweight='bold')
plt.xlabel('Customer Names',fontsize=15,fontweight='bold')
plt.ylabel('Discount Amt',fontsize=15,fontweight='bold')
plt.bar(discount['Customer_Name'][0:10],discount['Cash_Disc.Amt'][0:10],color = 'red')
```

```python
Time = cleaned_data[['Bill_Amount','Bill_No','BillTime']]
Time['BillTime'] = pd.to_datetime(Time['BillTime']).dt.round('H').dt.time
Time['BillTime'].value_counts()
Time.head()
```

```python
peak_Time = Time.groupby('BillTime')['Bill_Amount'].sum().reset_index()
peak_Time_values = peak_Time.sort_values('Bill_Amount',ascending=False)
peak_Time_values
```

```python
peak_hours = peak_Time[['BillTime','Bill_Amount']]
peak_hours
```

### Converting our values to string type so that we can visualize our time data using matplotlib

```python
time_list = []
for i in peak_hours['BillTime']:
    time_list.append(str(i))
print(time_list)
```

### Finding the peak hours of sales

```python
plt.figure(figsize=(15,7))
plt.plot(time_list,peak_hours['Bill_Amount'],marker='o',color = 'red')
plt.bar(time_list,peak_hours['Bill_Amount'],color = 'green')
plt.title('Peak Hours',fontweight='bold',fontsize=15)
plt.xlabel('Time')
plt.ylabel('sales Over Years')
```
